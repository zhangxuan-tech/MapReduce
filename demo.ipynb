```python
import json
import os
from tqdm import tqdm
def readJson(path):
    keys = []
    values = []
    isstoreKeys = False
    with open(path, 'r') as jsonfile:
        data = json.load(jsonfile)
        if len(data.keys()) > 1:
            print('出现错误！')
        for key in data.keys():
            for tc_dict in data[key]:
                if isstoreKeys == False:
                    # print(tc_dict.keys())
                    for k in tc_dict.keys():
                        keys.append(k)
                    isstoreKeys = True
                for v in tc_dict.values():
                    values.append(v)
    # print(keys)
    # print(values)
    return keys, values

def getTestCase(path):
    keyslist = []
    valueslist = []
    for file in os.listdir(path):
        if file.endswith('.txt.txt'):
            keys, values = readJson(os.path.join(path ,file))
            keyslist.append(keys)
            valueslist.append(values)
    return keyslist, valueslist

def write_string_to_file(file_path, keyslist, valueslist):
    key=''
    for k in keyslist[0]:
        key += str(k) + '#'
    key = key+'\n'
    keylen = len(keyslist[0])
    for values in valueslist:
        for i in range(len(values)):
            if (i+1) % keylen == 0:
                key += str(values[i]) + '#' + '\n'
            else:
                key += str(values[i]) + '#'

    try:
        # 使用 'w' 模式打开文件，该模式会覆盖文件原有内容
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(key)
        print(f"成功将内容写入文件：{file_path}")
    except Exception as e:
        print(f"写入文件时出错：{e}")



# file_path = '/data/muxy/reduceDataset/dataset/r186'
# keyslist, valueslist = getTestCase(file_path)
# write_string_to_file('/data/muxy/reduceDataset/De.txt', keyslist, valueslist)

path = '/data/muxy/reduceDataset/dataset'
for folder in tqdm(os.listdir(path)):
    file_path = os.path.join(path, folder)
    print(file_path)
    keyslist, valueslist = getTestCase(file_path)
    save_file = os.path.join(file_path, 'testcase.txt')
    write_string_to_file(save_file, keyslist, valueslist)
```

```python
def count_lines_in_file(file_path):
    """
    该函数用于计算指定文件的行数
    :param file_path: 要读取的文件的路径
    :return: 文件的行数，如果文件不存在则返回 0
    """
    line_count = 0
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for _ in file:
                line_count = line_count + 1
        return line_count
    except FileNotFoundError:
        print(f"未找到文件: {file_path}")
        return 0
```

```python
def count_hashes_in_first_line(file_path):
    try:
        # 以只读模式打开文件
        with open(file_path, 'r', encoding='utf-8') as file:
            # 读取文件的第一行
            first_line = file.readline()
            # 计算第一行中 '#' 的个数
            hash_count = first_line.count('#')
            return hash_count
    except FileNotFoundError:
        print(f"文件 {file_path} 未找到。")
        return 0
    except Exception as e:
        print(f"读取文件时发生错误: {e}")
        return 0
```

```python
import random

# 生成 0 到 20 之间的随机整数
def buchong_1(file_path, head_num, tail_num):
    cont = ''
    for i in range(tail_num):
        for j in range(head_num):
            cont += str(random.randint(0, 1000)) + '#'
        cont += '\n'
   

    
    try:
        # 使用 'w' 模式打开文件，该模式会覆盖文件原有内容
        with open(file_path, 'a', encoding='utf-8') as file:
            file.write(cont)
        print(f"成功将内容写入文件：{file_path}")
    except Exception as e:
        print(f"写入文件时出错：{e}")
```

```python
import random

def delete_random_five_lines_except_first(file_path, num_line):
    try:
        # 以读取模式打开文件并读取所有行
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()

        # 检查除第一行外的行数是否足够删除 5 行
        if len(lines) - 1 < num_line:
            print("除第一行外，文件剩余行数少于 5，无法删除 5 行。")
            return

        # 从第二行开始选取行的索引（因为第一行索引为 0，从 1 开始选取）
        available_indices = list(range(1, len(lines)))
        # 随机选择 5 个索引
        indices_to_delete = random.sample(available_indices, num_line)

        # 过滤掉要删除的行，保留第一行和其他不需要删除的行
        new_lines = [line for i, line in enumerate(lines) if i not in indices_to_delete]

        # 以写入模式打开文件，将新的行写入文件
        with open(file_path, 'w', encoding='utf-8') as file:
            file.writelines(new_lines)

        print("成功从文件中随机删除 5 行（第一行未删除）。")

    except FileNotFoundError:
        print(f"文件 {file_path} 未找到。")
    except Exception as e:
        print(f"处理文件时发生错误: {e}")
```

```python
def read_first_line(file_path):
    try:
        # 以只读模式打开文件，使用 UTF-8 编码
        with open(file_path, 'r', encoding='utf-8') as file:
            # 读取文件的第一行
            first_line = file.readline()
            # 如果第一行存在，去除行尾的换行符
            if first_line:
                first_line = first_line.rstrip('\n')
            return first_line
    except FileNotFoundError:
        print(f"文件 {file_path} 未找到。")
        return None
    except Exception as e:
        print(f"读取文件时发生错误: {e}")
        return None
```

```python
import os

def delete_files_starting_with(folder_path):
    # 检查文件夹是否存在
    if not os.path.exists(folder_path):
        print(f"指定的文件夹 {folder_path} 不存在。")
        return
    try:
        # 遍历文件夹中的所有文件和子文件夹
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            # 检查文件是否以 testcase_originaltest 开头且为文件而不是文件夹
            if filename.startswith('testcase_original') and os.path.isfile(file_path):
                # 删除文件
                os.remove(file_path)
                print(f"已删除文件: {file_path}")
    except PermissionError:
        print(f"没有权限删除文件夹 {folder_path} 中的文件。")
    except Exception as e:
        print(f"删除文件时出现错误: {e}")


# 示例调用
path = '/data/muxy/reduceDataset/dataset'
for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)
    delete_files_starting_with(folder_path)
```

```python
import shutil
import os

def delete_allset_folder(parent_dir):
    """
    此函数用于删除指定父目录下名为 AllSet 的文件夹
    :param parent_dir: 父目录的路径
    """
    allset_path = os.path.join(parent_dir, 'AllSet')
    if os.path.exists(allset_path) and os.path.isdir(allset_path):
        try:
            shutil.rmtree(allset_path)
            print(f"成功删除 {allset_path} 文件夹。")
        except Exception as e:
            print(f"删除 {allset_path} 文件夹时出现错误: {e}")
    else:
        print(f"{allset_path} 文件夹不存在。")



path = '/data/muxy/reduceDataset/dataset'
for folder in os.listdir(path):
    delete_allset_folder(os.path.join(path, folder))
```
